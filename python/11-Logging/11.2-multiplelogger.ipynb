```python
import logging

# Step 1: Configure global logging (file + format)
logging.basicConfig(
    filename='app.log',
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)

# Step 2: Create loggers for modules
logger1 = logging.getLogger('module1')
logger1.setLevel(logging.DEBUG)

logger2 = logging.getLogger('module2')
logger2.setLevel(logging.WARNING)

```

```python
# Step 3: Log messages
logger1.debug("This is a debug message from module1")
logger1.info("This is an info message from module1")

logger2.warning("This is a warning message from module2")
logger2.error("This is an error message from module2")

print("✅ Logs written to app.log")
```
