pydantic basics : creating and using models

pydantic models are the foundation of data validation in python they use puthon type annotation to definr the structure and validate data at runtime here a detailed exploration of base model creation and several example

```python
from pydantic import BaseModel
```

```python
from dataclasses import dataclass

@dataclass

class Person():
    name:str
    age:int
    city:str
```

```python
class Person(BaseModel):
    name:str
    age:int
    city:str

person=Person(name='osama',age=35,city="Bhopal")
print(person)
```

```python
class Person_(BaseModel):
    name:str
    age:int
    city:str

person=Person_(name='osama',age=35,city=0)
print(person)
```

```python
person1=Person(name="hehe",age=24,city=235)
```

```python
from typing import Optional
class Employee(BaseModel):
    id:int
    name:str
    department:str
    salary:Optional[float]=None
    is_active:Optional[bool]=True

```

```python
exp1=Employee(id=1,name="john",department='IT')
print(exp1)
```

```python
exp2=Employee(id=69,name="diddy",department='IT',salary=1000000,is_active=False)
print(exp2)
```

```python
from pydantic import BaseModel
from typing import List

class CLassroom(BaseModel):
    room_num:str
    students:list[str]
    capacity:int
```

```python
from pydantic import BaseModel, Field, field_validator


class Address(BaseModel):
    street: str = Field(min_length=3, description="Street name must be at least 3 characters long")
    city: str = Field(min_length=2)
    zip_code: str = Field(pattern=r"^\d{5}$", description="Zip code must be exactly 5 digits")

    # Custom validator to strip extra spaces
    @field_validator("city")
    def clean_city(cls, value):
        return value.strip().title()


class Customer(BaseModel):
    customer_id: int = Field(gt=0, description="Customer ID must be a positive number")
    name: str = Field(min_length=2)
    address: Address

    # Custom validator to format name
    @field_validator("name")
    def format_name(cls, value):
        return value.strip().title()


customer = Customer(
    customer_id=1,
    name='   diddy   ',
    address={'street': "123 main street", "city": " boston ", "zip_code": "01212"}
)

print(customer)
```
