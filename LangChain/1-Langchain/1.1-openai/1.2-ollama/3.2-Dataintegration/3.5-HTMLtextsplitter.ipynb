```python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_string = """
<html>
<body>
  <h1>Header 1</h1>
  <p>This is paragraph under header 1.</p>

  <h2>Header 2</h2>
  <p>Second section paragraph.</p>

  <h3>Header 3</h3>
  <p>Another section inside header 3.</p>
</body>
</html>
"""

header_to_splitter = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3"),
]

html_splitter = HTMLHeaderTextSplitter(headers_to_split_on=header_to_splitter)

html_header_splits = html_splitter.split_text(html_string)

print(html_header_splits)
```
